# Inter arrival time of calls

```python
data_frames = [df1, df2, df3, df4]
data_frames = copy.copy(data_frames)
new_data_frames = []
for df_i in data_frames:
    row_count = df_i.shape[0]
    start_time = 0
    for i in range(row_count):
        if i == 0:
            df_i.at[i, 'inter arrival'] = 0
            start_time = df_i.at[i, 'start time']
        else:
            df_i.at[i, 'inter arrival'] = df_i.at[i, 'start time'] - df_i.at[i-1, 'start time']
        df_i.at[i, 'Scheduler'] = df_i.at[i, 'Scheduler'] + " - "  + df_i.at[i, 'CPU stress']
        df_i.at[i, 'Scenario index'] = i
        
    new_data_frames.append(copy.copy(df_i[2:]))

big_df = pd.concat(new_data_frames, ignore_index=True)
print(big_df)
```

```python
import copy

measurement_num = 3

if measurement_num == 2:
    df_i = copy.copy(df2)
elif measurement_num == 1:
    df_i = copy.copy(df1)
elif measurement_num == 3:
    df_i = copy.copy(df3)
elif measurement_num == 4:
    df_i = copy.copy(df4)

    
#fine_tunes = [15000000+423511, -24900000+200000+1184896.0, 14400000, 0]
fine_tunes = [15000000+423511, -23914900, 14400001, 0]
DF2_START = 0
row_count = df_i.shape[0]
start_time = 0
deadline_miss = 0
period_slip = 0
for i in range(row_count):

    if i == 0:
        start_time = int(copy.copy(df_i.at[i, 'start time']))
        if measurement_num == 1:
            df_i.at[i, 'period start time'] = int(DF1_START) - start_time - fine_tunes[0]
        if measurement_num == 2:
            df_i.at[i, 'period start time'] = 0 + 184896
        if measurement_num == 3:
            df_i.at[i, 'period start time'] = int(DF3_START) - start_time - fine_tunes[2]
        if measurement_num == 4:
            df_i.at[i, 'period start time'] = 0
        df_i.at[i, 'start time'] = 0
        
    else:
        df_i.at[i, 'start time'] = int((df_i.at[i, 'start time'] - start_time))
        df_i.at[i, 'period start time'] = df_i.at[i-1, 'period start time'] + 450000000
    
    df_i.at[i, 'deadline'] = df_i.at[i, 'period start time'] + 450000000

    # DEADLINE MISS
    if df_i.at[i, 'deadline'] < df_i.at[i, 'start time'] + df_i.at[i, 'execution time']:
        deadline_miss += 1
        df_i.at[i, 'deadline miss'] = True
        #print("{}".format((df_i.at[i, 'start time'] + df_i.at[i, 'execution time'])-df_i.at[i, 'deadline']))
        print("DEADLINE MISS: {} < {}".format(df_i.at[i, 'deadline'], df_i.at[i, 'start time'] + df_i.at[i, 'execution time']))
        
        # reset period start
        df_i.at[i, 'period start time'] = df_i.at[i, 'start time'] + df_i.at[i, 'execution time']
    else:
        df_i.at[i, 'deadline miss'] = False
    
    #PERIOD SLIP
    if df_i.at[i, 'start time'] < df_i.at[i, 'period start time'] and i != 0:
        print("PERIOD SLIP: {} <? {}".format(df_i.at[i, 'start time'], df_i.at[i, 'period start time']))
        period_slip += 1
        df_i.at[i, 'period start time'] = df_i.at[i, 'start time'] - 1000000
    
    df_i.at[i, 'Scheduler'] = df_i.at[i, 'Scheduler'] + " - "  + df_i.at[i, 'CPU stress']
    df_i.at[i, 'Scenario index'] = i

print("DEADLINE MISS NUM: {}".format(deadline_miss))
print("PERIOD SLIP NUM: {}".format(deadline_miss))
```
